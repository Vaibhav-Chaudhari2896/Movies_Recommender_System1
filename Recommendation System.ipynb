## Defining functions for data cleaning

```python
# functions to extract genre names, cast names, crew names and keywords

def clean_genres(data):
    temp = []
    for i in ast.literal_eval(data):
        temp.append(i['name'].replace(" ",""))
    return temp
```

```python
def clean_cast(data):
    temp = []
    j = 0
    for i in ast.literal_eval(data):
        if j < 3:
            temp.append(i['name'].replace(" ",""))
            j += 1
    return temp
```

```python
def clean_crew(data):
    temp = []
    for i in ast.literal_eval(data):
        if i['job'] == 'Director':
            temp.append(i['name'].replace(' ',''))
            break
    return temp
```

```python
def clean_keywords(data):
    temp = []
    for i in ast.literal_eval(data):
        temp.append(i['name'].replace(' ',''))
    return temp
```

```python
# Applying the functions

movie_dataset.genres = movie_dataset.genres.apply(clean_genres)
movie_dataset.overview = movie_dataset.overview.apply(lambda x: x.split())
movie_dataset.cast = movie_dataset.cast.apply(clean_cast)
movie_dataset.crew = movie_dataset.crew.apply(clean_crew)
movie_dataset.keywords = movie_dataset.keywords.apply(clean_keywords)
```

```python
movie_dataset.head(2)
```

```python
# Joining all the features into single feature

movie_dataset['features'] = movie_dataset['genres'] + movie_dataset['overview'] + movie_dataset['cast'] + \
                            movie_dataset['keywords'] + movie_dataset['crew']
```

```python
# Dropping features 

movie_dataset.drop(columns=['genres', 'overview', 'cast', 'keywords', 'crew'], inplace=True)
```

```python
movie_dataset.head()
```

```python
movie_dataset.features = movie_dataset.features.apply(lambda x: ' '.join(x))
movie_dataset.features = movie_dataset.features.apply(lambda x: x.lower())
```

```python
movie_dataset.head()
```

```python
# Lemmatizing words to their base form inorder to extract more meaningful words

from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
# Defining stemmer function

def stem(data):
    temp = []
    for i in data.split():
        temp.append(ps.stem(i))
    return " ".join(temp)
```

```python
movie_dataset.features = movie_dataset.features.apply(stem)
```

```python
# Creating tfidf vectorizer

from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english', max_features = 10000)  # removes all words in stop_words
```

```python
# Transforming dataset to tfidf matrix

tfidf_matrix = tfidf.fit_transform(movie_dataset.features)
tfidf_matrix.shape
```

```python
# Checking similarity score for each movie

from sklearn.metrics.pairwise import cosine_similarity

similarity_score = cosine_similarity(tfidf_matrix, tfidf_matrix)
similarity_score.shape
```

```python
def recommend( movie, top = 5):
    movie_index = movie_dataset[movie_dataset['title'] == movie].index[0]
    ref = similarity_score[movie_index]
    movies_recommended = sorted(list(enumerate(ref)), reverse=True, key = lambda x: x[1]) [1:(top+1)]
    
    for i in movies_recommended:
        print(movie_dataset.iloc[i[0]].title)
    return
```
